```python
# importing libraries
# we use numpy, pandas, and math
import pandas as pd
import numpy as np
import math as m
```

```python
#importing dataset for fault diagnosis from sensor
reqmt = pd.read_excel('./dataset/need_maintenance.xlsx')
```

```python
# importing dataset
# this dataset containing all information that we need in this program
dataset = pd.read_excel('./dataset/dataset_workshop.xlsx') 
dataset.head()
```

```python
#searh Avaibility with case need battery maintenance
if reqmt.iloc[0]['battery'] == 'FAIL':
    dataset = dataset[dataset['battery']== 'Available']
if reqmt.iloc[0]['brakepad'] == 'FAIL':
    dataset = dataset[dataset['brakepad']== 'Available']
if reqmt.iloc[0]['motor'] == 'FAIL':
    dataset = dataset[dataset['motor']== 'Available']
dataset = dataset.reset_index()
data = dataset 
#we will process the 'dataset', so so we back up the 'dataset' into 'data'
workshop = dataset['workshop_name']
```

```python
dataset.head()
```

```python
#retrieve the measurement parameter
#we use distances, rating, and estimate time [collums 5,6,7]
dataset = dataset.iloc[:,5:8].values
row, col = dataset.shape
print(dataset)
print("Shape : {}x{}".format(row, col))
```

```python
# defining weights
# we setting weights value 30% for distance, 60% for ratings, and 10% for estimate time
w = [0.3, 0.6, 0.1]
```

```python
# Step 1 => calculating normalized matrix
sumOfSquares = []
for j in range(0, col):
    sum = 0
    for i in range(0, row):
        value = dataset[i][j]
        sum = sum + m.pow(value,2)
    sqrtOfSum = m.sqrt(sum)
    sqrtOfSum = round(sqrtOfSum, 4)
    sumOfSquares.append(sqrtOfSum)
print('Column-wise root of sum of squares', sumOfSquares)

# creating a new 2D array
dataset2 = [[0 for i in range(col)] for j in range(row)] 

# dividing by the respected values
for j in range(0, col):
    for i in range(0, row):
        temp = dataset[i][j]/sumOfSquares[j]
        dataset2[i][j] = round(temp, 4)

print('normalized matrix')
print(np.matrix(dataset2))
normalized = pd.DataFrame(np.matrix(dataset2))
normalized.columns =['distance_normalized_matrix', 'rating_normalized_matrix', 'time_normalized_matrix'] 
```

```python
# Step 2 => calculating weighted normalized matrix
for j in range(0, col):
    for i in range(0, row):
        temp2 = dataset2[i][j]*w[j]
        dataset2[i][j] = round(temp2, 4)
weighted = pd.DataFrame(np.matrix(dataset2))
weighted.columns =['distance_weighted_matrix', 'rating_weighted_matrix', 'time_weighted_matrix']
print('weighted normalized matrix')
print(np.matrix(dataset2))
```

```python
# Step 3 => calculating ideal best and ideal worst
# length of impacts should only increase if number of coulmns increase
impacts = ['-', '+', '-']
# as we know, the parameters for this scoring prosess are distances, rating, and estimate time
# the lower the ditances value, the better impact for the performance ('-')
# the higher the ditances rating, the better impact for the performance ('+')
# the lower the ditances estimate time, the better impact for the performance ('-')
bestValues = []
worstValues = []

dataset3 = pd.DataFrame(dataset2)
maxValues = dataset3.max()
minValues = dataset3.min()

for num in range(0, len(impacts)):
    if impacts[num] == '+':
        bestValues.append(maxValues[num])
        worstValues.append(minValues[num])
    else:
        bestValues.append(minValues[num])
        worstValues.append(maxValues[num])

# should only increase if number of columns increase
print('Ideal Best Values (r)\n', bestValues)
# print('')
print('Ideal Worst Values (r)\n', worstValues)
```

```python
# Step 4 => calculating euclidean distance from ideal bests and ideal worsts
SPlus = []
SMinus = []

# print(dataset2)

for i in range(0, row):
    bestSum = 0
    worstSum = 0
    for j in range(0, col):
        bestSum = bestSum + m.pow((dataset2[i][j]-bestValues[j]),2)
        worstSum = worstSum + m.pow((dataset2[i][j]-worstValues[j]), 2)
    SPlus.append(round(m.sqrt(bestSum), 4))
    SMinus.append(round(m.sqrt(worstSum), 4))
    
# should only increase if number of rows increase
plus_impact = pd.DataFrame(SPlus)
min_impact = pd.DataFrame(SMinus)
plus_impact.columns =['plus_impact']
min_impact.columns =['min_impact']
print('SPlus (c)\n', SPlus)
# print('')
print('SMinus (c)\n', SMinus)
```

```python
# Step 5 => calculating performance score
pi = []
for index in range(0, row):
    pi.append(round(SMinus[index] / (SMinus[index] + SPlus[index]), 4))
    
score = pd.DataFrame(pi)
score.columns =['score']
# should only increase if number of rows increase
print('Performance Score (c)\n', pi)
```

```python
#find maximum pi value 
max_value = max(pi)

#we need index information for maximum pi value 
max_index = pi.index(max_value)

#print the best workshop 
workshop[max_index]
```

```python
#create result dataset
result = pd.concat([data,normalized,weighted,plus_impact,min_impact,score], axis=1)
```

```python
#short values by score, and adding column rank
final_result = result.sort_values('score', ascending=False)
final_result['rank'] = range(1, 1+len(final_result))
```

```python
final_result.to_excel('final_result.xlsx')
```
